```python
from bs4 import BeautifulSoup
from pymongo import MongoClient
import requests
import json
from time import sleep
from random import randint
from warnings import warn
from time import time
from IPython.core.display import clear_output
import numpy as np
import pandas as pd
```

```python
areas = ['sandiego', 'inlandempire', 'dallas', 'newyork', 'sfbay', 'sanantonio', 'albuquerque', 'yuma', 'lasvegas', 'elpaso', 'bellingham', 'boise', 'lascruces', 'phoenix', 'buffalo', 'detroit', 'chicago', 'kansascity', 'stlouis', 'denver', 'reno', 'sacramento', 'stockton', 'seattle', 'losangeles',]
areas2 = ['washingtondc', 'baltimore', 'charlotte', 'atlanta', 'plattsburgh', 'miami', 'jacksonville', 'vermont', 'nh', 'philadelphia', 'pittsburgh', 'delaware', 'maine', 'bozeman', 'missoula', 'neworleans', 'shreveport', 'louisville', 'memphis', 'omaha', 'wichita', 'eugene', 'humboldt', 'slo', 'santabarbara', 'bakersfield', 'fresno', 'newhaven']
print(len(areas2))
for area in areas2:
    print(area)
```

```python
for area in areas2:
    orurl = 'https://' + area +'.craigslist.org/d/rooms-shares/search/roo'
    print(orurl)
    file = area + '.json'
    print(file)
    resp = requests.get(orurl)
    soup = BeautifulSoup(resp.text, 'html.parser')




#find the total number of posts to find the limit of the pagination
    results_num = soup.find('div', class_= 'search-legend')
    results_total = int(results_num.find('span', class_='totalcount').text)

    pages = np.arange(0, results_total, 120)

    iterations = 0

    times = []
    nbrhd = []
    titletexts = []
    links = []
    prices = []
    postdataraw =[]
    for page in pages:
    
        #get request
        r = requests.get(orurl + "?s=" + str(page))
        print(r.text[:500])
        sleep(randint(1,5))
     
        #throw warning for status codes that are not 200
        if r.status_code != 200:
            warn('Request: {}; Status code: {}'.format(requests, r.status_code))
        
        #define the html text
        soup = BeautifulSoup(r.text, 'html.parser')
    
        #define the posts
        posts = soup.find_all('li', class_= 'result-row')
        
    #extract data 
        for post in posts:

            if post.find('span', class_ = 'result-hood') is not None:

                #posting date
                #grab the datetime element 0 for date and 1 for time
                post_datetime = post.find('time', class_= 'result-date')['datetime']
                times.append(post_datetime)

                #neighborhoods
                post_nbrhd = post.find('span', class_= 'result-hood').text
                nbrhd.append(post_nbrhd)

            #    title text
                post_title = post.find('a', class_='result-title hdrlnk')
                post_title_text = post_title.text
                titletexts.append(post_title_text)

                 #post link
                post_link = post_title['href']
                links.append(post_link)

                post_price = post.a.text
                prices.append(post_price)
            
                #get raw data from post
                #sleep(randint(1,5))
                rurl = requests.get(post_link)
                postdataraw.append(rurl.text)


        
        print("Page: " + str(page))
    rooms = pd.DataFrame({"posted": times,
                     'neighborhood': nbrhd,
                     'title': titletexts,
                     'url': links,
                     'price': prices,
                     'roomdataraw': postdataraw})
    rooms.to_json(file, orient = 'table')

```
